```python
import glob
import pandas as pd
```

```python
lyrics = []
lyrics_text_list = []
idx = 0

for fn in glob.glob('./lyrics/*'):
    with open(fn) as lines:
        a = fn.split('/')[2]
        a = a.split('.')[0]
        a = a.split('-')
        cat = a[0]
        subcat = a[1]
        for line in lines:
            entry = dict(id=idx, cat=cat, subcat=subcat, text=line.strip())
            lyrics_text_list.append(line.strip())
            lyrics.append(entry)
            idx += 1

# Now you can access individual entries in the list using index, e.g., lyrics[0], lyrics[1], etc.
print(len(lyrics))
```

```python
import random
import time

# Your dataset as an empty list
lyrics = lyrics_text_list

# Function to display a hallucinated line
def display_hallucinated_line(hallucinate_level=0):
    if not lyrics:
        print("Lyrics dataset is empty.")
        return
    
    original_line = random.choice(lyrics)
    
    if hallucinate_level == 0:
        print(original_line.upper())
    else:
        words_in_line = original_line.split()
        num_words_to_replace = min(hallucinate_level, len(words_in_line))
        words_to_replace = random.sample(words_in_line, num_words_to_replace)
        
        new_line = [word if word not in words_to_replace else random.choice(words_in_line) for word in words_in_line]
        print(' '.join(new_line.upper()))

while True:
    display_hallucinated_line(hallucinate_level=0)
    time.sleep(5)
```
